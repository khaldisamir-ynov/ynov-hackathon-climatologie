```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Load raw data from current working directory
df = pd.read_csv("raw.csv", delimiter=";")

print("Shape:", df.shape)
print(df.head())

# Safety checks
for c in ["AAAAMM", "TN", "TX"]:
    if c not in df.columns:
        raise ValueError(f"Missing required column: {c}")

# Clean AAAAMM -> monthly Period
df["AAAAMM"] = (
    df["AAAAMM"]
    .astype(str)
    .str.replace(r"\.0$", "", regex=True)
    .str.zfill(6)
)

df["month"] = pd.to_datetime(
    df["AAAAMM"],
    format="%Y%m",
    errors="coerce"
).dt.to_period("M")

# Drop bad rows
df = df.dropna(subset=["month", "TN", "TX"]).copy()

# Make sure numeric
df["TN"] = pd.to_numeric(df["TN"], errors="coerce")
df["TX"] = pd.to_numeric(df["TX"], errors="coerce")

df = df.dropna(subset=["TN", "TX"]).copy()

# Add year
df["year"] = df["month"].dt.year.astype(int)

print("\nCleaned shape:", df.shape)
print("Years:", df["year"].min(), "->", df["year"].max())
print("Stations:", df["NUM_POSTE"].nunique() if "NUM_POSTE" in df.columns else "N/A")
df.head()
```
